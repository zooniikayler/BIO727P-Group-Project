### Webscraping the ChEMBL ID's

Now need to filter from the PKIDB the Chembl ID's which will be put into a DF with the INN_Name 
We are also adding a column for links to inhibitor image from CHEMBL Report Card website 
This will be used to generate images of inhibitors on our website 
Later this DF will be joined to the main Inhibitor DF

```python
#using the PKIDB database to get all the Kinase Inhinbitors details
import requests
r = requests.get('http://www.icoa.fr/pkidb/')
```

```python
#parse using BeautifoulSoup
from bs4 import BeautifulSoup
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
results = soup.find_all('tr')
#the tr html tag wraps all the key information about the kinases we need so seek this out
```

```python
print(results)
```

```python
print(results[0])
#the first entry of the html tr tags is the labels
```

```python
final_results = results[1:]
#we ignore the first set of tr html tags as these are the labels and go on from index 1 of the tr html tags to the end
print(final_results[0])
```

```python
first_result =results[1]
#we need to test on interation of the tr html tags to pull out the key information 
print(first_result)
```

```python
#We are pulling out the Chembl Id which is wrapped in the a (action) html tag so search for this
chembl = first_result.find_all('a')[1]
#I know that index 1 (canonically this is the second occurance) of the a html tag gives me the Chembl data
print (chembl)
type(first_result)
```

```python
import re
#now I need to use regex to pull out the key data
# I turn this a html tag details into a string
chembl = str(chembl)
print(chembl)
```

```python
id_chembl = re.search(r"(CHEMBL\d+)",chembl)
#use the regex above to seek out the CHEMBL key word group followed by a number of digits
# I need to identify CHEMBL as a key word as some entries in the PKIDB don't contain CHEMBL ID's so the 1st (canonically) second html tag will give us another database
# Group 2 is my actual numerical Chembl_ID which I pulled out using regex
id_chembl.group(0)
```

```python
#Make a for loop to loop over the whole construction and pull out key info

chembl_link = []

#This will iterate over the parsed html file and look for specific html tags and regex to pull out key info
for i in final_results:
    INN_Name = i.find('td').text
    chembl = i.find_all('a')[1]
    chembl = str(chembl)
    CHEMBL_ID = re.search(r"(CHEMBL\d+)",chembl)
    if CHEMBL_ID is not None:
        CHEMBL_ID = CHEMBL_ID.group(0)
    else:
        continue
        #we want to skip occurances where this is not identified 
    Image_link = "https://www.ebi.ac.uk/chembl/api/data/image/"+ CHEMBL_ID + ".svg?engine=indigo"
    
    #make this into a tuple with seperate parts
    chembl_link.append((INN_Name, CHEMBL_ID, Image_link))
```

I added the Image_link column this is generated from the CHEMBL Report Card website when searching by CHEMBL ID you will get a report card of that respective CHEMBL ID including an image of it's structure 
When the image is clicked you will be sent to an url of that image. 

e.g. "https://www.ebi.ac.uk/chembl/api/data/image/CHEMBL1794066.svg?engine=indigo"

I noticed that the only changing nature of this url link was the CHEMBL ID so after getting the respective CHEMBL ID's I feed them into this url structure and loop through them as shwon in the respective code above.

```python
#applying a tapular data structure 
#let's convert the tupules into a df

import pandas as pd

#make this into a dataframe with specific columns
chembl_id = pd.DataFrame(chembl_link, columns=['INN_Name', 'CHEMBL_ID', 'Image_link'])

#sort values alphabetically by name
chembl = chembl_id.sort_values(by='INN_Name')
chembl.head()

```
