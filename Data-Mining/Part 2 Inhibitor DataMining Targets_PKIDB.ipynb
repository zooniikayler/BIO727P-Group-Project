### Creating DF with Just respective Kinases

This DF will be added back to the main Inhibitor DF later so needs to have the INN_Name as well as to serve as an index 

```python
#We are going to pull out the Targets for each kinase inhibitor and place them in their own column

import pandas as pd
```

```python
df = pd.read_csv('PKIDB_CSV.csv')
```

```python
df
```

```python
CHEMBL = pd.read_csv("Chembl_id.csv")
#df["CHEMBL_ID"]=CHEMBL.CHEMBL_ID
```

```python
df
```

```python
#make a df with the seperate targets in seperate columns 

INN_Name = df.INN_Name
Targets = df.Targets

df = pd.DataFrame({'INN_Name': INN_Name, 'Targets': Targets})
kinase_targets = df.Targets.str.split("\n", expand=True,)
print(kinase_targets)
```

```python
#create dataframe with 
kinase_targets = kinase_targets.set_index([INN_Name])
```

```python
kinase_targets.columns = ["kinase_1","kinase_2", "kinase_3", "kinase_4",
                          "kinase_5", "kinase_6", "kinase_7", "kinase_8", 
                          "kinase_9", "kinase_10", "kinase_11", "kinase_12",
                          "kinase_13", "kinase_14", "kinase_15", "kinase_16",
                          "kinase_17", "kinase_18" ]
```

```python
kinase_targets
```

```python
kinase_targets = kinase_targets.join(CHEMBL.set_index('INN_Name'), on='INN_Name')
```

```python
#organise the order of my df 
kinase_targets= kinase_targets[["CHEMBL_ID", "kinase_1","kinase_2", "kinase_3", "kinase_4",
                          "kinase_5", "kinase_6", "kinase_7", "kinase_8", 
                          "kinase_9", "kinase_10", "kinase_11", "kinase_12",
                          "kinase_13", "kinase_14", "kinase_15", "kinase_16",
                          "kinase_17", "kinase_18"]]
```

```python
kinase_targets
```

```python
#get rid of rows with NaN in the CHEMBL_ID column
kinase_targets = kinase_targets[pd.notnull(kinase_targets['CHEMBL_ID'])]

```

```python
kinase_targets
```

```python
#Drop all NaN values in the kinase_1 column 
kinase_targets = kinase_targets[pd.notnull(kinase_targets['kinase_1'])]
```

```python
kinase_targets
```

```python
#Make it into a csv file
kinase_targets.to_csv("targets_PKIDB.csv")
```
