## Creating main Inhibitor DF

- We are adding the other Inhibitor csv files together to make a main Inhibitor DF
- we are adding the Chembl ID DF, the Kinase Targets DF and the main inhibitor DF (PKIDB) to make one with all the information

We got the main Inhibitor PKIDB inhibitor DataFrame from using the information used in http://www.icoa.fr/pkidb/ then copying and pasting this into Excel, making sure the correct columns and infromation is copied over then saving this as a CSV file i.e. PKIDB_CSV

```python
#We are creating the main PKIDB (PROTEIN KINASE INHIBITOUR DATABASE) Adding necesary columns and Taking
#Taking away useless columns
import pandas as pd
```

```python
#Read in the necessary csv files we will be using

#CHEMBL has key information about CHEMBL ID
CHEMBL = pd.read_csv("Chembl_id.csv")

#PKIDB will serve as the backbone of our main PKIDB
PKIDB = pd.read_csv("PKIDB_CSV.csv")

#Contains key info about Targets
Targets = pd.read_csv("targets_PKIDB.csv")
```

```python
CHEMBL
```

##### We need to seperate the Smiles column so that Smiles and InChi Keys are in seperate columns instead of in one column together

```python
#Create a df with the Smiles & InchiKey in seperate columns
INN_Name = PKIDB.INN_Name
Smiles= PKIDB.Canonical_Smiles_InChiKey
df = pd.DataFrame({"INN_Name": INN_Name, "Smiles": Smiles})



#Expand the Smiles column seperating by new line
#Now those 2 values are in different columns
Smiles = df.Smiles.str.split("\n", expand = True)

#Give the 2 columns their names
Smiles.columns = ["Smiles", "InChi_Key"]

#Include the column INN_Name from PKIDB.INN_NAME
Smiles["INN_Name"]= INN_Name
```

```python
Smiles
```

```python
#ADD Smiles data to the PKIDB Inhibitor main DF
PKIDB_inn = PKIDB.INN_Name
Smiles_inn = Smiles.INN_Name

#Drop Canonical_Smiles_InChiKey column in the PKIDB as we are now 
#adding back the seperate columns of Smiles & InChi-Key

PKIDB = PKIDB.drop(columns=["Canonical_Smiles_InChiKey"], axis =1)

#Add the Smiles df to the PKIDB df by index of thier INN_Names
BASE_PKIDB = Smiles.join(PKIDB.set_index("INN_Name"), on ="INN_Name")
```

```python
BASE_PKIDB
```

```python
#ADD the CHEMBL ID
CHEMBL_inn = CHEMBL.INN_Name

#add the Chembl ID to the PKIDB
CHEMBL = CHEMBL.join(BASE_PKIDB.set_index("INN_Name"), on="INN_Name")

#Drop the unwanted columns
Pre_final_PKIDB = CHEMBL.drop(columns = ['BrandName','Phase','Applicants','Links','LigID','pdbID','Type',
                                      'Indications','Targets','First_Approval','SC_Patent','Chirality',
                                      'FDA approved','Melting points (°C)','Synonyms',], axis=1)
```

```python
#add the img column to the pre_final_PKIDB
#Smiles["INN_Name"]= INN_Name

#Read in the img csv file 
img = pd.read_csv("Kinase_inhibiors_img_list.csv")

#add the column image_link into the pre_final_PKIDB DF
Pre_final_PKIDB["image_link"]= img.image_link
```

```python
Main_PKIDB = Pre_final_PKIDB.set_index("INN_Name")
Main_PKIDB
```

```python
#Get rid of the Unnamed column
Main_PKIDB = Main_PKIDB.loc[:, ~Main_PKIDB.columns.str.contains('unnamed', case=False)]
```
